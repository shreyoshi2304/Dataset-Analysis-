```python
import numpy as np
matrix = []
print("Enter a 3x3 matrix (integers only):")
for i in range(3):
    row = []
    for j in range(3):
        while True:
            try:
                val = int(input(f"Enter integer at position ({i+1},{j+1}): "))
                row.append(val)
                break
            except ValueError:
                print("Invalid input. Please enter an integer.")
    matrix.append(row)

matrix_np = np.array(matrix)
print("The Original 3x3 matrix is:\n",matrix_np)

matrix_multi = matrix_np * 2
print("The matrix multiplied by 2 is\n",matrix_multi)

matrix_add = matrix_np + 5
print("The matrix with 5 added to each element is:\n",matrix_add)

matrix_sq = matrix_np **2
print("The matrix with each element squared is:\n",matrix_sq)
```

```python
import numpy as np
matrix = []
print("Enter a 3x3 matrix (integers only):")
for i in range(3):
    row = []
    for j in range(3):
        while True:
            try:
                val = int(input(f"Enter integer at position ({i+1},{j+1}): "))
                row.append(val)
                break
            except ValueError:
                print("Invalid input. Please enter an integer.")
    matrix.append(row)

matrix_np = np.array(matrix)
print("The Original 3x3 matrix is:\n",matrix_np)

print("The first row of the array is:\n",matrix_np[0,:])
print("The last column of the array is:\n",matrix_np[:,2])

# I am assuming the question stating 2x2 matrix FROM centre means the first element of the 2x2 matix should the centre element

print("The 2x2 matrix from the centre is\n",matrix_np[1:3,1:3])
```

```python
import pandas as pd

name = []
marks = []

for i in range(10):
    while True:
        studentname = input("Enter student name: ")
        if not studentname.replace(" ", "").isalpha():
            print("Invalid name. Please enter letters only (no numbers or symbols).")
            continue
        studentmarks = input(f"Enter {studentname}'s marks: ")
        try:
            studentmarks = int(studentmarks)
        except ValueError:
            print("Invalid marks. Please enter an integer.")
            continue
        name.append(studentname)
        marks.append(studentmarks)
        break
data ={
       "NAME": name,
       "MARKS": marks
       }
data_set = pd.DataFrame(data)
print(data_set)
```

```python
import pandas as pd
name = []
salary = []

for i in range(5):
    while True:
        employeename = input("Enter employee name: ")
        if not employeename.replace(" ", "").isalpha():
            print("Invalid name. Please enter letters only (no numbers or symbols).")
            continue
        employeesalary = input(f"Enter {employeename}'s salary: ")
        try:
            employeesalary = int(employeesalary)
        except ValueError:
            print("Invalid marks. Please enter an integer.")
            continue
        name.append(employeename)
        salary.append(employeesalary)
        break
data ={
       "Employee_Name": name,
       "Income": salary
       }
data_set = pd.DataFrame(data,index =['a','b','c','d','e'])
print(data_set)
```
