```python
# Importing necessary modules
import time  # For adding delay between printed lines (used in typing effect)
from google import genai  # Google's generative AI SDK to access Gemini models
from google.colab import userdata  # To securely retrieve the API key stored in Colab

# Retrieve the API key securely from Colab's userdata
key = userdata.get("GeminiAPI")

# Attempt to initialize the client using the API key
try:
    client = genai.Client(api_key=key)
except KeyError:
    # Error handling in case the API key is not found
    print(f"There is an error with the API key.\nError message:\n{KeyError}")

# Create a chat session with the Gemini 2.5 Flash model
# I chose this model because it's lightweight, fast, and works well for conversational tasks
chat = client.chats.create(model="gemini-2.5-flash")

# Function to create a typing effect when displaying the chatbot's poetic response
# This makes the bot feel more alive and enhances user experience
def stream(text):
    print("Bot:")
    # Split the generated poem into lines (stanzas)
    lines = text.strip().split("\n")
    for line in lines:
        print(line)
        time.sleep(0.4)  # Small pause between lines to simulate natural delivery

# Prompt the user to select a favorite poet's style for the chatbot to mimic
# This adds personalization and thematic consistency to the poetic responses
poet = input("Enter your favourite poet (e.g., Rumi, Emily Dickinson, Shakespeare): ")

# Core function that interacts with the Gemini model
def ask_gemini(user_input):
    try:
        # Formulate a custom prompt that instructs the model to:
        # - Adopt the style of the selected poet
        # - Use lyrical and grammatically correct language
        # - Format the output into 4-line stanzas
        prompt = f"""
        Write a poetic response in the style of {poet} , it should represent the beauty and grandieur of human emotions and life.
        Use refined lyrical language, correct punctuation, and format the output into well-structured stanzas with four lines per stanza.
        Here is your query: {user_input}"""
        # Send the prompt to Gemini and get the model's poetic reply
        response = chat.send_message(prompt)
        return response.text
    except Exception as e:
        # Catch and display any errors that might occur during API call
        return f"Oops, something went wrong: {e}"

# Welcome message and loop to keep the chat running
print("Welcome to your Gemini-2.5-flash poetic bot! Remember to type 'end' to exit the chat.")

# Infinite loop to continuously interact with the user until they choose to exit
while True:
    user_input = input("You: ")
    if user_input.lower() == 'end':
        # When user chooses to quit, send a poetic goodbye
        print("Bot: With gratitude in every rhyme, farewell until another time.")
        break
    # Get Gemini's poetic reply and stream it with a natural flow
    response = ask_gemini(user_input)
    stream(response)
```
