```python
from transformers import pipeline
sentiment_analyzer = pipeline('sentiment-analysis')
model = "distilbert-base-uncased-finetuned-sst-2-english"
```

```python
sentence = "I love this movie."
result = sentiment_analyzer(sentence)
print(result)
```

```python
dataset = ["The poems are beautifully worded and flow well.","It lacks real interaction with user input.","The randomness adds surprise but not relevance.","The typing effect gives a nice human touch.","No AI model like Gemini is used, which was required.","It’s creative, but not truly intelligent.","Feels more like a poetry generator than a chatbot.","The program is simple and runs smoothly.","It doesn't demonstrate prompt engineering skills.","Good effort overall, but misses the main objective"]
results = sentiment_analyzer(dataset)
for sentence , result in zip(dataset,results):
    print(f"Sentence: {sentence}")
    print(f"Sentiment: {result['label']}, Score: {result['score']}")
```
