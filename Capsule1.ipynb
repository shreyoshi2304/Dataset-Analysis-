```python
import re
import numpy as np
valid_length_units = {
    'mm', 'millimeter', 'millimeters',
    'cm', 'centimeter', 'centimeters',
    'm', 'meter', 'meters',
    'in', 'inch', 'inches',
    'ft', 'foot', 'feet'
}
radius = input("Dear user please input a radius of your choice with appropriate units")
def extract_number_and_unit(text):
    # Match float or scientific notation followed by unit (singular or plural)
    pattern = r"""(?x)                           # verbose mode
        ^\s*                                     # optional leading space
        ([+-]?                                   # optional sign
         (?:\d+\.\d*|\.\d+|\d+)(?:[eE][+-]?\d+)?  # float or sci notation
        )\s*                                     # optional space
        ([a-zA-Z]+s?)\s*$                        # unit (with optional plural 's')
    """

    match = re.match(pattern, text)
    if match:
        number = float(match.group(1))
        unit = match.group(2).lower()  # lowercase for consistency
        return number, unit
    else:
        return None, None
radius_value, radius_unit = extract_number_and_unit(radius)

if radius_value is not None and radius_unit in valid_length_units:
    print(f"Radius value: {radius_value}")
    print(f"Radius unit: {radius_unit}")
    area = np.pi * (float(radius_value)**2)
    print ("The area of your cirle is {} {}^2".format(area,radius_unit))
else:
    print("Invalid input. Please enter a value like '2.5cm' or '1.2e-3 m'.")
```

```python
import re

current_year = int(input("Enter the current year: Just the year no need for adding AD "))
birth_year = input("Enter your birth year (e.g., '300 BC' or '1982 AD'): ")

def extract_number_and_timeline(text):
    pattern = r"""(?ix)                       # case-insensitive & verbose
        ^\s*
        ([+-]?(?:\d+\.\d*|\.\d+|\d+))         # numeric year
        \s*
        (BC|AD)                               # era (case-insensitive)
        \s*$
    """
    match = re.match(pattern, text)
    if match:
        number = float(match.group(1))
        timeline = match.group(2).upper()    # keep uppercase: "BC" or "AD"
        return number, timeline
    return None, None

birth_value, birth_timeline = extract_number_and_timeline(birth_year)


if birth_value is not None and birth_timeline in {"AD", "BC"}:
    birth_value = int(birth_value)

    if birth_timeline == "AD":
        if birth_value > current_year:
            print(" You are not born yet!")
        else:
            age = current_year - birth_value
            print(f"You are {age} years old.")
    elif birth_timeline == "BC":
        age = current_year + birth_value - 1
        print(f"You are {age} years old!! ARE YOU A VAMPIRE?!")
else:
    print("Invalid input. Please enter a year followed by 'BC' or 'AD' (e.g., '300 BC', '1982 AD').")
```

```python
import re
P = input("enter the principal amount")
R = input ("enter the rate of interest subject PER ANNUM")
T = input ("enter the time period in years or montths or both")
def convert_time_to_years(TT):
    TT= TT.lower().strip()
    # Regex patterns to match years and months
    year_match = re.search(r"([\d.]+)\s*(year|yr|yrs|years)", TT)
    month_match = re.search(r"([\d.]+)\s*(month|months)", TT)

    years = 0.0

    if year_match:
        years += float(year_match.group(1))

    if month_match:
        months = float(month_match.group(1))
        years += months / 12

    if years == 0:
        print("❌ Invalid input. Please enter something like '2 years', '18 months', or '1 yr 6 months'.")
        return None

    return round(years, 2)

converted_years = convert_time_to_years(T)

if converted_years is not None:
    print(f"Time in years: {converted_years}")

I = float(P)*(float(R)/100)*float(converted_years)
print("The simple interest of your sum is{} Rupees".format(I))
```
