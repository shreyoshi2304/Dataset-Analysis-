```python
from transformers import pipeline

# Optional: For better performance on certain models, you might need to import specific classes
# from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-fr")
```

```python
def chatbot():
    print("Welcome to the English-to-French Translator Chatbot!")
    while True:
        user_input = input("Enter English text (or type 'exit' to quit): ")

        if user_input.lower() == "exit":
            print("Goodbye!")
            break

        try:
            translation = translator(user_input)
            french_text = translation[0]['translation_text']
            print(f"French Translation: {french_text}")
        except Exception as e:
            print(f"An error occurred during translation: {e}")

if __name__ == "__main__":
    chatbot()
```

```python
   import time

   def chatbot():
       print("Welcome to the English-to-French Translator Chatbot!")
       while True:
           user_input = input("Enter English text (or type 'exit' to quit): ")

           if user_input.lower() == "exit":
               print("Goodbye!")
               break

           start_time = time.time()
           try:
               translation = translator(user_input)
               french_text = translation[0]['translation_text']
               end_time = time.time()
               translation_time = end_time - start_time
               print(f"French Translation: {french_text} (Translation took {translation_time:.2f} seconds)")
           except Exception as e:
               print(f"An error occurred during translation: {e}")

   if __name__ == "__main__":
       chatbot()
```
